```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a 3D vector
pv = [-0.86978817, 0.47860026, 0.12004302]
ov = [-0.97850794, 0.040889535, -0.20211479]
pv_97 = [-0.91548973, 0.36909294, 0.1601528]
ov_97 = [-0.9822093, 0.18357803, 0.0395477]
#ov_97 = [0.9822093, 0.38357803, 0.1395477]
v = np.array(pv_97)
v2 = np.array(ov_97)

def draw(v, v2): # red is pointing, blue is obj
    # Normalize the vector to get the unit vector
    unit_vector = v / np.linalg.norm(v)
    unit_vector2 = v2 / np.linalg.norm(v2)

    # Plotting the unit vector in 3D
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the vector
    ax.quiver(0, 0, 0, unit_vector[0], unit_vector[1], unit_vector[2], length=1, color='r')

    ax.quiver(0, 0, 0, unit_vector2[0], unit_vector2[1], unit_vector2[2], length=1, color='b')

    # Set limits for the axes
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])

    # Set labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.set_title(f"pointing 3D Unit Vector: {unit_vector}, obj {unit_vector2}")

    # Show the plot
    plt.show()
    
#draw(pv, ov)
draw(pv_97, ov_97)
```
